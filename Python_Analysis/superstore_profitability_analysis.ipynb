# Superstore Profitability Analysis

This notebook explores the **Superstore Sales dataset** to understand how the business is performing and where profit is coming from.

The goal of this analysis is simple:

- Understand sales and profit behavior
- Identify the most profitable categories and regions
- Find areas where the business is losing money

This type of analysis is very common for data analysts working with retail or e‑commerce companies.

## 1. Import Libraries

First we import the Python libraries that help us work with data.

```python
import pandas as pd
import numpy as np
```

## 2. Load the Dataset

We load the Superstore dataset using pandas.

Make sure the CSV file is in the same folder as this notebook or provide the correct path.

```python
df = pd.read_csv('superstore_cleaned.csv')
df.head()
```

## 3. Understanding the Data

Before doing analysis, it is important to understand the dataset structure.

We check:

- Column names
- Data types
- Number of rows
- Missing values

```python
df.info()
```

```python
df.describe()
```

## 4. Total Sales and Total Profit

One of the first things a business wants to know is the total sales and total profit.

```python
total_sales = df['sales'].sum()
total_profit = df['profit'].sum()

print(total_sales, total_profit)
```

## 5. Profit Margin

Profit margin tells us what percentage of sales becomes profit.

Formula:

Profit Margin = Profit / Sales

```python
profit_margin = (total_profit / total_sales)*100
print(profit_margin)
```

## 6. Sales by Category

This helps us understand which product categories generate the most revenue.

```python
category_sales = df.groupby('category')['sales'].sum().sort_values(ascending=False)
category_sales
```

## 7. Profit by Category

Sales alone does not tell the full story. Some categories may sell a lot but still produce low profit.

```python
category_profit = df.groupby('category')['profit'].sum().sort_values(ascending=False)
category_profit
```

## 8. Regional Performance

Companies also want to know which regions perform best.

```python
region_profit = df.groupby('region')['profit'].sum().sort_values(ascending=False)
region_profit
```

## 9. Most Profitable Products

This helps identify which individual products generate the most profit.

```python
top_products = df.groupby('product_name')['profit'].sum().sort_values(ascending=False).head(10)
top_products
```

## 📊 Key Insights

The analysis of the Superstore sales dataset reveals that high sales do not always lead to high profit. In many cases, large discounts reduce profitability and even create losses. Some products and sub-categories generate strong revenue but have very low or negative profit margins.

Another important observation is that profitability varies across regions and product categories. Certain categories contribute significantly to revenue, while others maintain healthier profit margins despite lower sales.

These patterns highlight that sales growth alone is not enough; profit-focused decision making is essential for sustainable business performance.

## 💡 Recommendations

To improve profitability, the business should control excessive discounting, especially on products that already have low margins. Loss-making products should be carefully reviewed and either repriced or removed from the catalog.

The company should also focus more on high-profit categories and products, while applying region-specific sales strategies based on performance patterns. By prioritizing profitable products and optimizing pricing strategies, the Superstore can reduce losses and improve overall profitability.
